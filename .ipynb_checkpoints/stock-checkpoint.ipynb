```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import seaborn as sns
import requests
```

```python
r = requests.get('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=MSFT&apikey=3MRYI5QDD949CMEO')
# r.json()['Time Series (Daily)']
r.json()
#call this api to get MS stock daily price in json format
```

```python
tt = r.json()['Time Series (Daily)']
#get only the daily price part, remove Information part
```

```python
t1 = pd.DataFrame.from_dict(tt)
#make json to DataFrame(pandas dataformat)
```

```python
t2 = t1.T
#make it index by date column
```

```python
t3 = t2[::-1]
#reverse the column, start from old date to new date
t3
```

```python
short_rolling = t3.rolling(window=11).mean()
short_rolling
#get 11 days sma value
```

```python
ema_short = t3.ewm(span=20, adjust=False).mean()
ema_short
#get 20 days ema value
```

```python
start_date = '2018-10-03'
end_date = '2019-02-12'


fig, ax = plt.subplots(figsize=(15,9))

ax.plot(t3.loc[start_date:end_date, :].index, t3.loc[start_date:end_date, '4. close'], label='Price')
#draw daily close price line, index by very day
ax.plot(ema_short.loc[start_date:end_date, :].index, ema_short.loc[start_date:end_date, '4. close'], label = 'EMA')
#draw 20 days ema value line based on daily close price, index by very day
ax.plot(short_rolling.loc[start_date:end_date, :].index, short_rolling.loc[start_date:end_date, '4. close'], label = 'SMA')
#draw 11 days sma value line based on daily close price, index by very day

ax.legend(loc='best')
ax.set_ylabel('Price in $')

ax.xaxis.set_major_formatter(my_year_month_fmt)
```
